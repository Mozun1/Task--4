```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("traffic-accident-statistics-as-of-1439-h.csv")

plt.figure(figsize=(8,5))

plt.title('traffic accident statistics as of 1439')

plt.plot(df.الشهر , df.السبت ,'r.-')
plt.plot(df.الشهر , df.الاثنين ,'m.-')
plt.plot(df.الشهر , df.الثلاثاء ,'b.-')
plt.plot(df.الشهر , df.الخميس ,'g.-')


plt.xticks(df.الشهر[::2])

plt.xlabel('month')
plt.ylabel('Total accidents')



plt.show()
```
